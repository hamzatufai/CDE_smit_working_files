```python
for i in range(10):
    print(i)
    break
```

```python
for i in range(10):
    print(i)

    if i == 5:
        print("Breaking loop")
        break
    
```

```python
for i in range(10):
    if i == 0:
        print("Continue")
        continue 

    print(i)

    
```

```python
for i in range(10):
    if i == 5 or i == 7:
        # print("Continue")
        continue 
    print(i)
   
```

### While loop

```python
# while condition:
#   code statement
```

```python
while True: # Infinte iterations (always true )
    print(1)
```

```python
num = 0 

while num < 10:
    num = num + 1 # num += 1 
    print(num)
```

```python
num = 0
while num < 10:
    num = num + 1 # num += 1 
    print(num)

    if num == 8:
        break
```

```python
retries = 0
max_retries = 3
connected = False

while not connected and retries < max_retries:
    print(f"Attempting connection... (attempt {retries + 1})")
    retries += 1
    # if retries == 2:
    #     connected = True

if connected:
    print("Connection established.")
else:
    print("Connection failed after maximum retries.")
```

```python
# Nested Loops 

# ***** 
# *****
# *****
# *****
# *****

for i in range(5):
    # print('*')
    for j in range(5):
        print('*', end='')
    print()

# *
# **
# ***
# ****
# *****
```

```python
for i in range(5):  # 0,1,2,3,4
    for j in range(5): # 0,1,2,3,4
        print(i,j)
    # print()
    # break
```

```python
tools = ["AWS", "Azure", "PySpark", "DBT", "Snowflake","Databricks"]

for tool in tools:
    print(tool)

    for char in tool:
        print(char)

# 1st loop 

# 1st iteration "AWS"   ---  2nd loop --- total 3 iterations  (a, w, s)
# 2nd iteration  "Azure"   ---  2nd loop --- total 5 iterations  (A, z,u,r,e)
```
