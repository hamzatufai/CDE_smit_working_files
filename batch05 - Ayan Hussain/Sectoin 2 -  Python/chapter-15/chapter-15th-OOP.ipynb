```python
class BankAccount: # class name 
    """A simple bank account."""

    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance

    def deposit(self, amount): # method
        self.balance += amount
        print(f"Deposited amount {amount} and new balance is {self.balance}")

    def withdraw(self, amount): # method
        self.balance -= amount
        print(f"withdraw amount {amount} and new balance is {self.balance}")
    
    def balance_check(self):
        print(f"Current balance is {self.balance}")
```

```python
haseeb = BankAccount("haseeb", 10000) # class object 
mohsin = BankAccount("mohsin", 20000) # class object 

print(haseeb.owner)
print(haseeb.balance)
```

# Inheritance

```python
class Animal:
    def __init__(self, name):
        self.name = name

    def eat(self):
        return f"{self.name} is eating."

    def speak(self):
        return f"{self.name} makes a sound."

class Dog(Animal):

    def speak(self):     # overrides             
        return f"{self.name} says: Woof!"

    def bark(self):
        return f"{self.name} barks."
    
dog = Dog("german_shephard")
print(dog.name)
dog.eat()
dog.speak()
```

```python
class Vehicle:
    def __init__(self, brand, speed,model):
        self.brand = brand 
        self.speed = speed
        self.model = model
    
    def info(self):
        print(f'{self.brand}, {self.speed},{self.model} ')

# duplication without using super 

class Car(Vehicle):
    def __init__(self,brand, speed,model,doors):
        self.brand = brand 
        self.speed = speed
        self.model = model
        self.doors = doors
    
class Truck(Vehicle):
    def __init__(self,brand, speed,model,tires):
        self.brand = brand 
        self.speed = speed
        self.model = model
        self.tires = tires
```

```python
class Vehicle:
    def __init__(self, brand, speed,model):
        self.brand = brand 
        self.speed = speed
        self.model = model
    
    def info(self):
        print(f'{self.brand}, {self.speed},{self.model} ')


class Car(Vehicle):
    def __init__(self, brand, speed,model,doors ):
        super().__init__(brand, speed,model)
        self.doors = doors
    
class Truck(Vehicle):
    def __init__(self,brand, speed,model,tires):
        super().__init__(brand, speed,model)
        self.tires = tires
```
