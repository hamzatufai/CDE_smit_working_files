```python
tools = ["AWS", "Azure", "PySpark", "DBT", "Snowflake","Databricks"]

if "AWS" in tools:
    print("We will learn AWS")
else:
    print("Not Exists")
```

```python
if "GCP" in tools: # False
    print("We will learn AWS")
else:
    print("Not Exists")
```

```python
if "GCP" in tools: # False
    print("We will learn AWS")
    print(123)

elif "Snowflake" in tools:
    print("Snowflake")

elif "Databricks" in tools:
    print("Snowflake")

else:
    print("Not Exists")
```

```python
if "Snowflake" in tools and "Databricks" in tools:
    print("Yes both exists")
```

```python
if "Snowflake" in tools and "GCP" in tools:
    print("Yes both exists")
else:
    print("No")
```

```python
if "Snowflake" in tools or "GCP" in tools:
    print("Yes both exists")
else:
    print("No")
```
