```python
data = {
    "name":   ["Alice", "Bob", "Carol", "David"],
    "salary": [85000, None, 91000, 65000],
    "city":   ["Karachi", None, "Lahore", None]
}
df = pd.DataFrame(data)
df
```

```python
# print(df.isnull())          # True where value is missing
print(df.isnull().sum())    # count of missing per column
```

```python
# Fill salary with mean, drop rows missing city
df["salary"] = df["salary"].fillna(df["salary"].mean())
df_clean = df.dropna(subset=["city"])
print(df_clean)
```

```python
df
```

```python
df = df.rename(columns={
    "name":   "full_name",
    "salary": "annual_salary"
})
print(df.columns)
```

```python
df["annual_salary"] = df["annual_salary"].astype(int)
print(df.dtypes)
```

```python
def salary_band(annual_salary):
    if annual_salary >= 90000:
        return "Senior"
    elif annual_salary >= 75000:
        return "Mid"
    else:
        return "Junior"

df["band"] = df["annual_salary"].apply(salary_band)
print(df[["full_name", "annual_salary", "band"]])
```

```python
df
```

```python
# Sort by salary descending
print(df.sort_values("annual_salary", ascending=True)[["full_name", "annual_salary"]])
```

```python
summary = df.groupby("department")["salary"].mean()
print(summary)
```

```python
summary = df.groupby("department")["salary"].agg(
    count="count",
    mean="mean",
    max="max"
)
print(summary)
```
