# Dictionaries

```python
# key-value pairs 

client = {
    # key:value
    "name": "Danial",
    "industry": "Finance",
    "projects": 3,
    "active": True
}

print(client)
```

```python
client['name']
```

```python
client['projects']
```

```python
client['inactive']
```

```python
# get()

print(client.get('name')) # returns value of key
```

```python
print(client.get('inactive', 'Not-Available')) # by default returns None
```
